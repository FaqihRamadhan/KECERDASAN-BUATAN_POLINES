C.MENDAPATKAN NILAI QUARTILE DAN INTERQUARTILE RANGE

```python
import numpy as np
data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]
Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1
Q1, Q3, IQR
```

```python
import numpy as np

# Data hasil ujian
nilai_ujian = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

# 1. Urutkan data
nilai_ujian.sort()

# 2. Identifikasi Q1, Q3, dan IQR
q1 = np.percentile(nilai_ujian, 25)
q3 = np.percentile(nilai_ujian, 75)
iqr = q3 - q1

# 3. Tentukan Upper Bound dan Lower Bound
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# 4. Tentukan Outlier
outlier = [x for x in nilai_ujian if x < lower_bound or x > upper_bound]

print(f"Data Terurut: {nilai_ujian}")
print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")
print(f"Batas Bawah: {lower_bound}, Batas Atas: {upper_bound}")
print(f"Data Outlier: {outlier}")
```

```python
import numpy as np

# Menentukan titik tengah (xi) dari setiap rentang berat badan
# 41-45(43), 46-50(48), 51-55(53), 56-60(58), 61-65(63), 66-70(68)
titik_tengah = np.array([43, 48, 53, 58, 63, 68])
frekuensi = np.array([2, 3, 6, 4, 5, 2])

# 1. Hitung Mean (Rata-rata) data kelompok
n = np.sum(frekuensi)
mean_kelompok = np.sum(titik_tengah * frekuensi) / n

# 2. Hitung Standar Deviasi
# Rumus: akar dari ( sigma f * (xi - mean)^2 / (n - 1) )
varians = np.sum(frekuensi * (titik_tengah - mean_kelompok)**2) / (n - 1)
std_deviasi = np.sqrt(varians)

print(f"Rata-rata Berat Badan: {mean_kelompok:.2f} kg")
print(f"Standar Deviasi: {std_deviasi:.2f}")
```
