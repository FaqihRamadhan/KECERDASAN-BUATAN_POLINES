**3. MENGGUNAKAN FUNGSI MATH**

PENUGASAN

1.PERKALIAN MATRIX 

```python
import numpy as np

# Definisi Matriks
A = np.array([[3, 1], [2, 0]])
B = np.array([[2, 1], [1, -1]])
C = np.array([[-1, 1], [2, 2], [0, -1]])

# A x B (Perkalian Matriks)
hasil_AB = np.matmul(A, B) # atau A @ B
print("Hasil A x B:\n", hasil_AB)

# B . C
# Catatan: Matriks B (2x2) dan C (3x2) tidak dapat dikalikan secara standar (B x C) 
# karena jumlah kolom B tidak sama dengan jumlah baris C. 
# Namun jika yang dimaksud adalah C x B:
hasil_CB = np.matmul(C, B)
print("Hasil C x B:\n", hasil_CB)
```

2.HITUNG NILAI FAKTORIAL

```python
import math

# Menghitung 6!
faktorial_6 = math.factorial(6)
print("Nilai 6! =", faktorial_6)

# Menghitung 8! / 5!
faktorial_8_5 = math.factorial(8) / math.factorial(5)
print("Nilai 8! / 5! =", faktorial_8_5)
```

3.OPERASI VEKTOR 3D

```python
import numpy as np

vec_A = np.array([5, 3, 7])
vec_B = np.array([12, -3, 1])

# Panjang masing-masing Vektor (Magnitude)
panjang_A = np.linalg.norm(vec_A)
panjang_B = np.linalg.norm(vec_B)

# Hasil Perkalian Titik (Dot Product)
dot_product = np.dot(vec_A, vec_B)

# Hasil Perkalian Silang (Cross Product)
cross_product = np.cross(vec_A, vec_B)

print(f"Panjang Vektor A: {panjang_A}")
print(f"Panjang Vektor B: {panjang_B}")
print(f"A . B (Dot): {dot_product}")
print(f"A x B (Cross): {cross_product}")
```
